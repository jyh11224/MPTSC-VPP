```python
df_real = df_sort[df_sort['구분1']=='실측']
df_est = df_sort[df_sort['구분1']=='예측']
```

```python
df_real = df_real.loc[:, ['일자','발전소용량(KW)', '구분1', '발전소코드', '합계']]
df_est = df_est.loc[:, ['일자','발전소용량(KW)', '구분1', '발전소코드', '합계']]
```

```python
merged = (pd.merge(df_real, df_est, left_on=["발전소코드","일자"], right_on=["발전소코드","일자"]))
merged
```

```python
import math
def sub(a, b):
  return abs(a - b)

merged['오차'] = merged.apply(lambda x: sub(x['합계_y'], x['합계_x']), axis=1)
```

```python
merged
```

```python
merged['nMAE_오차율'] = merged['오차'] / merged['발전소용량(KW)_x'] * 100
```

```python
merged
```

```python
def conditions(merged):
    if (merged['nMAE_오차율'] <= 6 ):
        val = 4
    elif (merged['nMAE_오차율'] <= 8):
        val = 3
    else:
        val = 0
    return val
```

```python
merged['기준']= merged.apply(conditions, axis=1)
```

```python
merged
```

```python
merged['reward'] = merged['합계_x'] * merged['기준']
```
